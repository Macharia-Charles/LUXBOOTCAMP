#### Import pandas library

```python
import pandas as pd
```

#### Reading the downloaded datasets

```python
# read csv files
_2018Data = pd.read_csv('IT Salary Survey EU 2018.csv')
_2019Data = pd.read_csv("T Salary Survey EU 2019.csv")
_2020Data = pd.read_csv("IT Salary Survey EU  2020.csv")
```

#### EDA Techniques
##### 1. Descriptive statistics.
Using the _'describe()'_ method to get a summary of the descriptive statistics of the datasets.

```python
# Get summary statistics for numerical columns
_2018Data.describe()
_2019Data.describe()
_2020Data.describe()
```

##### 2. Finding missing data.
Using the _'isnull()'_ method to check for missing values in the dataset. You can then use the sum() method to count the number of missing values in each column.

```python
# Check for missing values
_2018Data.isnull().sum()
_2019Data.isnull().sum()
_2020Data.isnull().sum()
```

##### 3. Distribution of categorical variables
Using the _'value_counts()'_ method to get a count of the unique values in a categorical column. You can then use this to create a bar chart to visualize the distribution.

```python
# Get the count of unique values in a categorical column
_2018Data['Gender'].value_counts()
_2019Data['Gender'].value_counts()
_2020Data['Gender'].value_counts()

# Visualize the distribution of the 'Gender' column
_2018Data['Gender'].value_counts().plot(kind='bar')
```

```python
_2019Data['Gender'].value_counts().plot(kind='bar')
```

```python
_2020Data['Gender'].value_counts().plot(kind='bar')
```

##### 4. Distribution of Numerical Variables
Using the _'hist()'_ method to create a histogram to visualize the distribution of a numerical column.

```python
_2018Data['Current Salary'].hist(bins=10)
```

```python
_2019Data['Yearly brutto salary (without bonus and stocks)'].hist(bins=10)
```

```python
_2020Data['Yearly brutto salary (without bonus and stocks) in EUR'].hist(bins=20)
```

###### 5. Correlation
Using the _'corr()'_ method to calculate the correlation between numerical columns in the dataset. You can then use a heatmap to visualize the correlation matrix.

```python
# Calculate the correlation matrix
corr1 = _2018Data.corr()
corr2 = _2019Data.corr()
corr3 = _2020Data.corr()

# Visualize the correlation matrix using a heatmap
import seaborn as sns
sns.heatmap(corr1, cmap='coolwarm', annot=True)
```

```python
sns.heatmap(corr2, cmap='coolwarm', annot=True)
```

```python
sns.heatmap(corr3, cmap='coolwarm', annot=True)
```
